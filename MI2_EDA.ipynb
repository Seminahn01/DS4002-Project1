```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load dataset
file_path = "/movie_data.csv"
df = pd.read_csv(file_path)
```

```python
# drop unnecessary column (ignore if column doesn't exist )
df = df.drop(columns=['Unnamed: 0'], errors='ignore')
```

```python
# Create a new column for review length
df['review_length'] = df['review_text'].apply(lambda x: len(str(x)))
```

```python
# Plot: Number of Reviews per Movie (Top 20 Movies)
top_movies = df['title'].value_counts().head(50)
plt.figure(figsize=(12, 6))
top_movies.plot(kind='bar', edgecolor='black')
plt.xlabel('Movie Title')
plt.ylabel('Number of Reviews')
plt.title('Top 20 Movies with Most Reviews')
plt.xticks(rotation=90)
plt.show()
```

```python
# Group by title and calculate the average lifetime gross per movie
avg_lifetime_gross = df.groupby('title')['lifetime_gross'].mean().sort_values(ascending=False)
top_20_avg_lifetime_gross = avg_lifetime_gross.head(50)
plt.figure(figsize=(12, 6))
top_20_avg_lifetime_gross.plot(kind='bar')

plt.xlabel("Movie Title")
plt.ylabel("Average Lifetime Gross ($)")
plt.title("Top 20 Movies by Average Lifetime Gross")
plt.xticks(rotation=90)
plt.show()
```

```python
# Scatter Plot: Review Length vs. Lifetime Gross
plt.figure(figsize=(10, 5))
plt.scatter(df['review_length'], df['lifetime_gross'], alpha=0.5)
plt.xlabel('Review Length (characters)')
plt.ylabel('Lifetime Gross ($)')
plt.title('Scatter Plot of Review Length vs. Lifetime Gross')
plt.xlim(0, df['review_length'].max())
plt.ylim(0, df['lifetime_gross'].max())
plt.grid(True)
plt.show()
```
