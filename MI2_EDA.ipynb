```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load dataset
file_path = "/movie_data.csv"
df = pd.read_csv(file_path)
```

```python
# drop unnecessary column (ignore if column doesn't exist )
df = df.drop(columns=['Unnamed: 0'], errors='ignore')
```

```python
# Create a new column for review length
df['review_length'] = df['review_text'].apply(lambda x: len(str(x)))
```

```python
# Plot: Number of Reviews per Movie (Top 20 Movies)
top_movies = df['title'].value_counts().head(50)
plt.figure(figsize=(12, 6))
top_movies.plot(kind='bar', edgecolor='black')
plt.xlabel('Movie Title')
plt.ylabel('Number of Reviews')
plt.title('Top 20 Movies with Most Reviews')
plt.xticks(rotation=90)
plt.show()
```

```python
# Scatter Plot: Review Length vs. Lifetime Gross
plt.figure(figsize=(10, 5))
plt.scatter(df['review_length'], df['lifetime_gross'], alpha=0.5)
plt.xlabel('Review Length (characters)')
plt.ylabel('Lifetime Gross ($)')
plt.title('Scatter Plot of Review Length vs. Lifetime Gross')
plt.xlim(0, df['review_length'].max())
plt.ylim(0, df['lifetime_gross'].max())
plt.grid(True)
plt.show()
```

```python
# Scatter plot: Review Year vs. Release Year
df_cleaned = df.dropna(subset=['review_year', 'release_year'])
df_cleaned['review_year'] = df_cleaned['review_year'].astype(int)
df_cleaned['release_year'] = df_cleaned['release_year'].astype(int)

plt.figure(figsize=(10, 6))
plt.scatter(df_cleaned['review_year'], df_cleaned['release_year'], alpha=0.5)
plt.xlabel("Review Year")
plt.ylabel("Release Year")
plt.title("Relationship between Review Year and Release Year")
plt.grid(True)
plt.show()
```

```python
# Scatter plot: Release year vs. lifetime gross
df_cleaned = df.dropna(subset=['release_year', 'lifetime_gross'])
df_cleaned['release_year'] = df_cleaned['release_year'].astype(int)
df_cleaned['lifetime_gross'] = df_cleaned['lifetime_gross'].astype(float)

plt.figure(figsize=(10, 6))
plt.scatter(df_cleaned['release_year'], df_cleaned['lifetime_gross'], alpha=0.5)
plt.xlabel("Release Year")
plt.ylabel("Lifetime Gross (in dollars)")
plt.title("Relationship between Release Year and Lifetime Gross")
plt.grid(True)
plt.show()
```
